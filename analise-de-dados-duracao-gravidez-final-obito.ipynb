```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

# Análise de Dados: Óbitos por nível de gravidez no período de 2016, relação parto e duração na capital de São Paulo.

## Leitura do arquivo  Excel, ao abrir podemos comparar a quantidade de tabelas com as tabelas apresentadas abaixo:

```python
df=pd.read_csv('./Dados/duracao-gravidez.csv', sep=';' , header=3, encoding='ANSI')
```

```python
df.head()
```

## Para deixar as tabelas mais bonitas, eliminamos as duas últimas colunas e a seguir podemos ver os níveis de óbito pela duração de parto.

```python
df=df[:3]
df.head()
```

```python
df.sort_values(by=['Menos de 22 semanas'],ascending=False).head(1)
```

```python
df.sort_values(by=['42 semanas e mais'],ascending=False).head(2)
```

## Ao gerar o gráfico podemos relacionar cada tipo de parto com sua duração. É possivel perceber que tem um grande número  de crianças perdidas antes de seu nascimento no período de 37 a 41 semanas comparado com 22 a 27 semanas.

```python
df.plot(kind = 'bar', x = 'Óbito relação parto', grid = True, title = 'Casos de óbitos por duração de gravidez', figsize = (15,5))
plt.ylabel('Quantidade')
```

## Antes do parto morreram mais crianças do que depois do parto as 22 semanas de gravidez.

```python
df.plot(kind = 'bar', y = 'Menos de 22 semanas', x = 'Óbito relação parto', grid = True, figsize = (15,5), title='Óbitos com menos de 22 semanas')
```

## Com 22 até 27 semanas cerca de quase 1200 crianças vieram a óbito antes do parto. Tem um baixo número de óbito a durante o parto comparado com o número de ignorados.

```python
df.plot(kind = 'bar', y = '22 a 27 semanas', x = 'Óbito relação parto', grid = True, figsize = (15,5), title='Óbitos com 22 a 27 semanas')
```

## Com apenas 28 a 31 semanas o nível durante o parto é baixissimo comparado com antes do parto...

```python
df.plot(kind = 'bar', y = '28 a 31 semanas', x = 'Óbito relação parto', grid = True, figsize = (15,5), title='Óbitos com 28 a 31 semanas')
```

## Quase chegando aos 8 meses, mais de 1200 bebês vieram a óbito antes do parto

```python
df.plot(kind = 'bar', y = '32 a 36 semanas', x = 'Óbito relação parto', grid = True, figsize = (15,5), title='Óbitos com 32 a 36 semanas')
```

## Quase aos 9 meses de gestação, quase mil vieram a óbito antes do parto...

```python
df.plot(kind = 'bar', y = '37 a 41 semanas', x = 'Óbito relação parto', grid = True, figsize = (15,5), title='Óbitos com 37 a 41 semanas')
```

## O número  total de ignorados em relação a cada parto:

```python
df.plot(kind = 'bar', y = 'Ignorado', x = 'Óbito relação parto', grid = True, figsize = (15,5), title='Óbitos Ignorados')
```
